### Manipulando para o formato wide (mais amigável)

```python
concept_wide = pd.merge(concept, concept_ref, on='CONCEPT_ID', how='inner')
concept_wide.drop(columns=['CONCEPT_ID'], inplace=True)
concept_wide = concept_wide.pivot(index=['DATE_REF', 'TRADINGITEM_ID'], columns='CONCEPT_NAME', values='CONCEPT_VALUE')
concept_wide
```

```python
concept_wide.columns
```

```python
indicator_wide = pd.merge(indicator, indicator_ref, on='INDICATOR_ID', how='inner')
indicator_wide.drop(columns=['INDICATOR_ID'], inplace=True)
indicator_wide = indicator_wide.pivot(index=['DATE_REF', 'TRADINGITEM_ID'], columns='INDICATOR_NAME', values='INDICATOR_VALUE')
indicator_wide
```

```python
indicator_wide.columns
```

```python
borrowrates.loc[borrowrates.loc[:, 'MARKET']=='Balcao', 'MARKET'] = 'BALCAO'
borrowrates_wide = borrowrates.pivot(index=['DATE_REF', 'TRADINGITEM_ID'], columns='MARKET', values=['CONTRACT_COUNT', 'SHARE_COUNT', 'VOLUME', 'DONOR_MIN', 'DONOR_MEAN', 'DONOR_MAX', 'TAKER_MIN', 'TAKER_MEAN', 'TAKER_MAX'])
borrowrates_wide
```

```python
borrowrates_wide.columns
```

```python
marketdata_wide = marketdata.set_index(['DATE_REF', 'TRADINGITEM_ID'])
marketdata_wide
```

```python
borrowrates_wide.index
```

```python
all_wide = pd.concat([concept_wide, indicator_wide, borrowrates_wide, marketdata_wide], axis=1, join='outer')
all_wide
```

```python
# Filtrando por uma data específica no nível 'DATE_REF'
all_wide.loc['2010-11-16']
```

```python
# Filtrando por um TRADINGITEM
all_wide.xs(342899, level='TRADINGITEM_ID')
```
